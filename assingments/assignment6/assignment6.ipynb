(a) Compute function for each individual (map)

<font size = "3">

- Import the datset "data/bmi.csv"
    - **Note that the column `Height` is in $cm$, and `Weight` is in $kg$**
- Write a lambda function for $ BMI = \frac{weight (kg)}{height (m)^2} $
- Use ```list(map(...))``` to calculate $BMI$ for each row of the dataset
- Round the outcome into 2 decimal points
- Store the calculation in a new column in the dataframe called `BMI`
- Display the dataframe


````python
# Write your own code here
data_bmi= pd.read_csv("data/bmi.csv")
func1 = lambda height, weight: weight/(height/100)**2
data_bmi["BMI"]= list(map(func1,data_bmi["Height"],data_bmi["Weight"]))
func2 = lambda bmi: round(bmi,2)
data_bmi["BMI"]= list(map(func2, data_bmi["BMI"]))
display(data_bmi)
````

- Print the original `countryname` and the newly created column, like: <br>
"['Afghanistan', 'Africa Eastern and Southern', 'Africa Western and Central', 'Albania', 'Algeria', ......] "

(d) Subset elements with query

<font size = "3">

- Compute the 25 quantile value of "prop_urbanpopulation"
- Use ".query()" to subset countries with <br>
 "prop_urbanpopulation" higher the 25 quantile

**hint: Import ```numpy``` and use `np.nanpercentile`**

(e) Plot by category

<font size = "3">

- Read "data/wdi_urban.csv"<br>
This contains data from 1980 and 2020

- Create a list with the unique values of "year"
- Create a plot with a ...
    - Histogram of "prop_urbanpopulation" <br>
    for each year
    - Set the histogram option ``` alpha = 0.5 ``` for transparency
    - All the histograms should be overlapping
    - Create a legend with the list of years
    - Label the axes and the title

What do we learn from the graph?

HINT: See the example at the end of Lecture 10


````python
# Write your own code here
data_wdi_urban=pd.read_csv("data/wdi_urban.csv")
year_unique=data_wdi_urban["year"].unique()
for year in year_unique:
    data_year = data_wdi_urban[data_wdi_urban["year"]==year]
    plt.hist(data_year["prop_urbanpopulation"], alpha = 0.5)
plt.xlabel("Urban Population")
plt.ylabel("Frequencies")
plt.legend(year_unique)
plt.title("The Distribution of Urban Population in Different Years")
plt.show()
print("From the graph, we can learn that generally the urban population increases from 1980 to 2020")
    


````
